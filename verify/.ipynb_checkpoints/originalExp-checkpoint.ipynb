### 加载相应模块

```python
###加载相应的模块
import numpy as np
import matplotlib.pyplot as plt
import cv2
import math
import numpy as np
from collections import Counter
import pandas as pd
import sys
import logging##使用日志显示一些输出
```

### 第一种文件读取方法，数字中间是空格

```python
####创建二维数组表示每个点的坐标
def Read1(arg1):
    coord=[]#### 构建二维数组
    for line in open(arg1,"r"):
        items=line.strip("\n").split(" ")
        tmp=[]##构建一维数组
        for item in items:
            tmp.append(float(item))
        coord.append(tmp)

    coord=np.array(coord)
    length=len(coord)##表示坐标的个数
    ##生成了所有点的坐标
    return coord,length
```

### 第二种文件读取方法，数字中间是逗号，并且含有正确的label

```python
def Read2(arg1):
    coord = []
    label = []
    for line in open(arg1, "r"):
        items = line.strip("\n").split(",")
        label.append(int(items.pop()))
        tmp = []
        for item in items:
            tmp.append(float(item))
        coord.append(tmp)
    coord = np.array(coord)
    label = np.array(label)
    length = len(coord)
    return coord,length
```

### 计算任意两点之间的距离

```python
####计算任意两个点之间的距离
def Distance(length,coord):
    distance=np.ones((length,length),dtype=float)
    dis=[]##表示所有距离列表
    for i in range(length):
        for j in range(i+1,length):
            dd=np.linalg.norm(coord[i]-coord[j])
            dis.append(dd)
            distance[i][j]=dd
            distance[j][i]=dd
    return distance,dis
```

### 设置阈值dc

```python
###开始设置点的阈值
def Threshold(percent,dis):
    percent = 2.0
    position = int(len(dis) * percent / 100)
    sortDis = np.sort(dis)
    dc = sortDis[position] #阈值
    return dc
```

### 计算每一个点的密度

```python
###根据阈值确定每一个点的密度

def Rho(distance,dc,length):
    rho=np.zeros(length)
    for i in range(length):
        count=0
        for j in range(length):
            ###包括自己
            if distance[j][i]<dc:
                count=count+1
        rho[i]=count
    return rho  
###函数不能和返回值重名
```

### 计算每一个点距离密度比它大的点的最小距离delta并且把该点记录下来

```python
###根据密度确定每一个点对应的delta，以及每个点对应的索引
def Delta(rho,distance,length):
    rho_index=np.argsort(-rho)
    ##寻找每一个i对应的比i密度大，距离又比较小的j作为i的中心
    ### 密度最大的点的中心是自己
    close_leader2=np.ones(shape=len(rho_index),dtype=np.int32)
    close_leader2[rho_index[0]]=rho_index[0]##这种情况不能进入for循环，需要提前写出来
    delta=np.zeros(len(rho_index))
    for i in range(1,len(rho_index)):
        minDis=distance[rho_index[0],rho_index[i]]
        index=rho_index[0]
        ##全部初始化为第一个点和i之间的距离
        for j in range(0,i):
            dis=distance[rho_index[i],rho_index[j]]
            if dis<minDis:
                minDis=dis
                index=rho_index[j]
        close_leader2[rho_index[i]]=index
        delta[rho_index[i]]=minDis
    ###开始寻找距离最大密度点的最大距离
    maxDis=distance[rho_index[0],rho_index[1]]
    delta[rho_index[0]]=maxDis
    for i in range(1,length):
        dd=distance[rho_index[0],rho_index[i]]
        if dd>maxDis:
            maxDis=dd
            index2=rho[i]
    delta[rho_index[0]]=maxDis
    return delta,close_leader2
```

### 绘制rho-delta决策图

```python
####可以画图选择簇心
#### 根据rho以及delta选出簇心
def show_decision1(rho, delta):
    plt.figure(num=1, figsize=(7, 5))
    for i in range(len(rho)):
        plt.scatter(x=rho[i], y=delta[i], c='k', marker='o', s=10)
    plt.xlabel('rho')
    plt.ylabel('delta')
    plt.title(r'Decision Graph ')
    plt.show()
```

### 开始选择簇心

```python
#w=np.array(rho)*np.array(delta)##对应乘积
```

```python
####开始确定簇心的索引
def choseList(w):
    leaders_num = int(input('input clusters num'))  # 根据以上的数据处理,输入最终选择的司令数目
    chose_list = np.argsort(-w)[: leaders_num]  # 选择一下这些点，选出最大的几个
    chose_list=list(chose_list)### 如果不加上这一个结果会很诡异
    return chose_list
###返回的是簇心的索引
```

```python
# 确定类别点,计算每点的密度值与最小距离值的乘积，并画出决策图，以供选择将数据共分为几个类别
def show_nodes_for_chosing_mainly_leaders(w):
    w=-np.sort(-w)[:]
    plt.figure(num=1, figsize=(8, 5))
    plt.scatter(x=range(len(w)), y=-np.sort(-w), c='k', marker='o')
    plt.xlabel('data_num')
    plt.ylabel('w')
    plt.title('Guarantee The Leader')
    plt.show()
```

### 再次绘制rho-delta图，并且将簇心使用不同的颜色表示出来，观察是不是理想的簇心

```python
def show_decision2(rho,delta,chose_list):
    plt.figure(num=1, figsize=(7, 5))
    for i in range(len(rho)):
        if i not in chose_list:
            plt.scatter(x=rho[i], y=delta[i], c='k', marker='o', s=10)
        else:
            plt.scatter(x=rho[i], y=delta[i], c='r', marker='o', s=10)
    plt.xlabel('rho')
    plt.ylabel('delta')
    plt.title(r'Decision Graph ')
    plt.show()
```

### 按照迭代的方式按所有点归入簇心

```python
def closeLeader2(close_leader2,chose_list,length):
    ####找到了每个点最近的leader之后，根据chose_list将每一个点迭代归入簇心
    close_lead=np.ones(length,dtype=int)
    for i in range(length):
        j=i
        while close_leader2[j] not in chose_list:
            j=close_leader2[j]
        close_lead[i]=close_leader2[j]
    return close_lead
```

```python
####画图显示
def show_result3(close_lead,coord,chose_list):
    colors = ['#BC8F8F','#4169E1','#8B4513','#FA8072','#FAA460',
              '#2E8B57','#FFF5EE','#A0522D','#C0C0C0', '#87CEEB','#6A5ACD', '#708090',
              '#FFFAFA','#00FF7F','#4682B4','#D2B48C','#D8BFD8','#FF6347','#40E0D0',
              '#EE82EE','#F5DEB3','#FFFFFF','#F5F5F5','#FFFF00',]
              
    
    leader_color = {}
    for index, i in enumerate(chose_list):
        leader_color[i] = index
        ####i的颜色是index
    plt.figure( figsize=(8,7))
    for node, class_ in enumerate(close_lead):
        #  标出每一类的聚类中心点
        if node not in chose_list:
            plt.scatter(coord[node, 0],coord[node, 1],c=colors[leader_color[class_]], s=10, marker='o', alpha=1)
    
    ###再将簇心绘制在上面
    for i in chose_list:
        plt.scatter(coord[i,0],coord[i,1],c="k",s=100,marker="+",alpha=1)
    
    plt.title('The Result Of Cluster')
    plt.show()
```

### 代码汇总

```python
def main1():
    arg=input("name of the file:")
    coord,length=Read1(arg)
    ##dis表示所有的距离形成的列表
    distance,dis=Distance(length,coord)
    percent=input("value of the precent:")
    dc=Threshold(percent,dis)
    rho=Rho(distance,dc,length)
    delta,close_leader2=Delta(rho,distance,length)
    show_decision1(rho, delta)###第一个决策图
    w=np.array(rho)*np.array(delta)##对应乘积
    show_nodes_for_chosing_mainly_leaders(w)##乘积对应的图
    chose_list=choseList(w)
    print("the chose_list:",chose_list)
    show_decision2(rho,delta,chose_list)##第二个决策图，标出了簇心
    close_lead=closeLeader2(close_leader2,chose_list,length)###所有点归入簇心
    show_result3(close_lead,coord,chose_list)##最终显示
```
